## Data Cleaning

```python
#data merge, removal of duplicate/unnecessary columns
cols_to_use = childCareInspec.columns.difference(childCareCenters.columns)
shortChildCareInfo = pd.merge(childCareCenters, childCareInspec[cols_to_use], left_index=True, right_index=True, how='outer')
shortChildCareInfo = shortChildCareInfo.drop(columns = ['URL'])
shortChildCareInfo.shape
```

```python
newViolationCat= shortChildCareInfo.groupby(['Violation Category'])
newViolationCat.size().sort_values()
```

```python
shortChildCareInfo[shortChildCareInfo['Violation Category']==""] = np.NaN
shortChildCareInfo['Violation Category'] = shortChildCareInfo['Violation Category'].fillna('OTHER')
newViolationCat= shortChildCareInfo.groupby(['Violation Category'])
newViolationCat.size().sort_values()
```

```python
shortChildCareInfo = shortChildCareInfo.dropna(how='any',axis=0) 
shortChildCareInfo.shape
```

## Data Visualization <br> 
Here it can be seen that Brooklyn is in the lead with the most instances. Brooklyn also falls in second place when it comes to the percent of eligible children served 

```python
ax = sns.countplot(y = 'Borough',
                                     data = shortChildCareInfo,
                                     facecolor = (0, 0, 0, 0), 
                                     linewidth = 5, 
                                     edgecolor = sns.color_palette('Set2'))
ax.set_ylabel('Frequents of Boroughs in NYC')
ax.set_xlabel('Frequency')
ax = ax.set_title('Instances')

```

```python
ax = sns.countplot(x = 'Borough',
                                     data = shortChildCareInfo,
                                     hue = 'Violation Category',
                                     palette = sns.color_palette("Set2"))

ax.set_xticklabels(ax.get_xticklabels(), rotation = 40, ha = "right")
water_legend = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax.set_ylabel('Frequents of Boroughs in NYC')
ax.set_xlabel('Frequency')
ax = ax.set_title('Instances')


```
